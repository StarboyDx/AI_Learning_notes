```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import math
import warnings

plt.rcParams['font.sans-serif'] = ['SimHei']
plt.rcParams['axes.unicode_minus'] = False
warnings.filterwarnings('ignore')
```

# 多图形绘制
## subplot

```python
# 1. 准备画布
plt.figure(1, figsize = (12, 6), facecolor = 'lightblue')
# 设置画布的编号和前景色

# 2. 画布分割
plt.subplot(1,2,1)
# 把画布分割成一行两列，定位到位置1axes上

# 3. 在指定的axes上进行绘制
# 3.1 左侧位置
plt.plot([7, 3, 2])
# plt会自动创建一个 和y对应的0开始的整数数组构成坐标进行绘制
plt.title('画布1, 区域1')

# 3.2 右侧位置
ax2 = plt.subplot(1,2,2)
# 沿用之前的分割，定位到第二个axes区域，并将该axes对象返回给调用函数的变量
ax2.bar([1,2,3],[5,15,9])
ax2.set_title('画布2, 区域2')

# 4. 构建一个新画布进行绘制
plt.figure(2, figsize=(12,6), facecolor='lightpink')
# 构建一个新画布，并定位

# 5. 绘制新的图形
plt.plot([100,200,80])
plt.title('画布2')

# 6. show
plt.show()
```

## subplots (返回的是figure和一个axes数组，可以用数组操作定位)

```python
# 1. 准备画布并进行切分
figure1, axes1 = plt.subplots(2, 2, figsize=(10,10))
# print(axes1)
# print(type(axes1))

# 2. 准备数据
x = range(8)
plt.gcf().suptitle('画布切分案例', y = 0.92)
# gcf: get_current_figure
# suptitle()副标题
# plt.title('大图')
# 直接使用title会定位到最后一个axes

# 3. 图形绘制
# 3.1 在第一个axes上绘制
axes1[0,0].plot(x, np.random.randint(1, 10, 8))
axes1[0,0].set_title('画布1, 区域1')

# 3.2其他区域绘制
axes1[0,1].bar(x, np.random.randint(1, 10, 8))

axes1[1,0].barh(x, np.random.randint(1, 10, 8))

axes1[1,1].scatter(x, np.random.randint(1, 10, 8))

# 4. show
plt.show()
```

```python
# 雇员数据集的可视化
df_emp = pd.read_excel('data/pandas.xlsx', 'Sheet1')
df_emp
```

```python
# 男性和女员工在销售额上的差异
df_emp.groupby(['Gender']).Sales.sum()
```

```python
df_emp.groupby(['Gender']).Sales.sum().plot(kind = 'bar')
```

```python
df_emp.groupby(['Gender']).Sales.mean().plot(kind = 'barh')
```

从上图可以看出男女平均销售额没显著差异

```python
df_emp.groupby('Gender').Sales.count().plot(kind = 'pie')
```

```python
# 不同性别，不同体型的员工销售额差异
df_emp.groupby(['Gender', 'BMI']).Sales.mean()
```

```python
sgb = df_emp.groupby(['Gender', 'BMI']).Sales.mean()
sgb.plot(kind = 'bar', figsize = (12,6), rot = 0)
```

具有层次化索引的Series直接plot得到的效果，是将不同层次的index组合成x轴的坐标，然后和对应的数据进行绘制

```python
sgb.unstack()
# unstack是把靠近数据的index旋转到columns上
```

```python
sgb.unstack().plot(kind = 'bar', figsize = (12,6), rot = 0)
```

从上图可以看出：男性员工不同体型下差异不大，女性员工差异较大

```python
sgb.unstack('Gender')
# 将指定的index旋转成为column
```

```python
sgb.unstack('Gender').plot(kind = 'bar', figsize = (12,6), rot = 0)
```

```python
sgb.unstack('Gender').plot(kind = 'bar', figsize = (12,6), rot = 0, stacked = True)
```
