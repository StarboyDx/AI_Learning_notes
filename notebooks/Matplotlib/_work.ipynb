```python
import pandas as pd
import matplotlib.pyplot as plt

plt.rcParams['font.sans-serif'] = ['SimHei']
plt.rcParams['axes.unicode_minus'] = False
```

```python
s1 = pd.Series([1900,1800,1450,1150,750,550,450,400,300,250,150],
               index = ['广州','其他','上海','北京','深圳','合肥','成都','杭州','武汉','南京','西安'])
```

```python
s1.plot(kind = 'bar', rot = 0)
plt.title('职位数按城市分布情况')
```

```python
# 补充方法
fig1=plt.figure(figsize=(12,6))
ax1=plt.subplot(1,1,1)

city = ['广州', '其他', '上海', '北京', '深圳', '合肥', '成都', '杭州', '武汉', '南京', '西安']
job = [1900, 1800, 1400, 1100, 750, 600, 400, 350, 250, 200, 150]
colors = ['#1f77b4', '#ff7f0e', '#2ca02c', '#d62728', '#9467bd', '#8c564b', '#e377c2', '#7f7f7f', '#bcbd22', '#17becf', '#1f77b4']
# 绘制柱状图
rects=ax1.bar(x=city,
             height=job,
             color=colors,
             width=0.5)
# 设置图表标题和坐标轴
ax1.set_ylim(0,2000)
ax1.set_xlabel('city',fontsize=20)
ax1.set_ylabel('job_counts',fontsize=20)
ax1.set_title('职位数按城市分布情况')
plt.show() 
```
