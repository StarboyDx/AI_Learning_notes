```python
#通过js形式将数据全部显示到提示框中
tool_js = """function (param) {return param.seriesName + ' — ' 
+param.data[3]+'<br/>' 
            +'人均GDP： '+param.data[0]+' 美元<br/>'
            +'GDP总量： '+param.data[2]+' 美元<br/>'
            +'人均寿命： '+param.data[1]+'岁';}"""

# 配色方案直接从Echarts投过来就好 
item_color_js_1 = """new echarts.graphic.RadialGradient(0.4, 0.3, 1, [{
                                        offset: 0,
                                        color: 'rgb(251, 118, 123)'
                                    }, {
                                        offset: 1,
                                        color: 'rgb(204, 46, 72)'
                                    }])"""
# 第二个配色方案
item_color_js_2 = """new echarts.graphic.RadialGradient(0.4, 0.3, 1, [{
                                        offset: 0,
                                        color: 'rgb(129, 227, 238)'
                                    }, {
                                        offset: 1,
                                        color: 'rgb(25, 183, 207)'
                                    }])"""

# 这个JS函数会根据GDP总量的数据计算一个数值，用于配置图形大小
symbol_js = """function (data) {return Math.sqrt(data[2]) / 5e2;}"""

# 绘制1990数据的散点图（必须公用一个x轴的数据才能一起添加，这两个年份的x轴数据（人均GDP）显然是不一样的）
scatter1 = (Scatter()
           .add_xaxis([i[0] for i in data[0]])
           .add_yaxis("1990年", [[i[1], i[2], i[3]] for i in data[0]],
                       symbol_size=JsCode(symbol_js), # 这里配置图形大小，根据GDP总量计算出symbol_size
                       label_opts=opts.LabelOpts(is_show=False), #关闭标签显示
                       itemstyle_opts=opts.ItemStyleOpts(color0='rgba[25,100,150,0.5]', color=JsCode(item_color_js_1)) #设置图形颜色
           )
           .set_global_opts(yaxis_opts=opts.AxisOpts(name='人均寿命', type_="value", is_scale=True), #is_scale设置Y轴缩放
                            xaxis_opts=opts.AxisOpts(name='人均GDP', type_="value"),
                            tooltip_opts = opts.TooltipOpts(formatter=JsCode(tool_js))
           )
)
# 2015数据绘制 (scatter1中配置了全局配置，是一样的这里就不用加了)
scatter2 = (Scatter()
           .add_xaxis([i[0] for i in data[1]])
           .add_yaxis("2015年", [[i[1], i[2], i[3]] for i in data[1]], 
                      symbol_size=JsCode(symbol_js), # 这里配置图形大小，根据GDP总量计算出symbol_size
                      label_opts=opts.LabelOpts(is_show=False),
                      itemstyle_opts=opts.ItemStyleOpts(color=JsCode(item_color_js_2)))
           )
scatter1.overlap(scatter2)           
scatter1.render_notebook()
```

```python
#通过js形式将数据全部显示到提示框中
tool_js = """function (param) {return param.seriesName + ' — ' 
+param.data[3]+'<br/>' 
            +'人均GDP： '+param.data[0]+' 美元<br/>'
            +'GDP总量： '+param.data[2]+' 美元<br/>'
            +'人均寿命： '+param.data[1]+'岁';}"""

# 配色方案直接从Echarts投过来就好 
item_color_js_1 = """new echarts.graphic.RadialGradient(0.4, 0.3, 1, [{
                                        offset: 0,
                                        color: 'rgb(251, 118, 123)'
                                    }, {
                                        offset: 1,
                                        color: 'rgb(204, 46, 72)'
                                    }])"""
# 第二个配色方案
item_color_js_2 = """new echarts.graphic.RadialGradient(0.4, 0.3, 1, [{
                                        offset: 0,
                                        color: 'rgb(129, 227, 238)'
                                    }, {
                                        offset: 1,
                                        color: 'rgb(25, 183, 207)'
                                    }])"""

# 这个JS函数会根据GDP总量的数据计算一个数值，用于配置图形大小
symbol_js = """function (data) {return Math.sqrt(data[2]) / 5e2;}"""

#ItemStyleOpts其实是没包含阴阳参数配置的，不过对于Pyecharts中的参数其实都支持直接传入如下字典形式来配置
# 图元样式配置，通过字典传入，包含阴影的设置
item_style_1 = {
    'shadowBlur': 10,
    'shadowColor': 'rgba(120, 36, 50, 0.5)',
    'shadowOffsetY': 5,
    'color': JsCode(item_color_js_1)
}
item_style_2 = {
    'shadowBlur': 10,
    'shadowColor': 'rgba(120, 36, 50, 0.5)',
    'shadowOffsetY': 5,
    'color': JsCode(item_color_js_2)
}

# 借用echarts的配色方案
bg_color_js = """
new echarts.graphic.RadialGradient(0.3, 0.3, 0.8, [{
        offset: 0,
        color: '#f7f8fa'
    }, {
        offset: 1,
        color: '#cdd0d5'
    }])"""

# 绘制1990数据的散点图（必须公用一个x轴的数据才能一起添加，这两个年份的x轴数据（人均GDP）显然是不一样的）
scatter1 = (Scatter(init_opts=opts.InitOpts(bg_color=JsCode(bg_color_js), #配置背景
                    # 长宽设置
                    width='1000px', height='800px'))
           .add_xaxis([i[0] for i in data[0]])
           .add_yaxis("1990年", [[i[1], i[2], i[3]] for i in data[0]],
                       symbol_size=JsCode(symbol_js), # 这里配置图形大小，根据GDP总量计算出symbol_size
                       label_opts=opts.LabelOpts(is_show=False), #关闭标签显示
                       itemstyle_opts=item_style_1 #设置图形颜色
           )
           .set_global_opts(yaxis_opts=opts.AxisOpts(name='人均寿命', type_="value", is_scale=True), #is_scale设置Y轴缩放
                            xaxis_opts=opts.AxisOpts(name='人均GDP', type_="value",
                                                     splitline_opts=opts.SplitLineOpts(is_show=True, #分割线，下面一行线性配置里设置为dashed
                                                     linestyle_opts=opts.LineStyleOpts(type_='dashed'))), 
                            tooltip_opts = opts.TooltipOpts(formatter=JsCode(tool_js)),
                            legend_opts=opts.LegendOpts(is_show=True, pos_right=10),
                            title_opts=opts.TitleOpts(title="1990 与 2015 年各国家人均寿命与 GDP")
           )
)
# 2015数据绘制 (scatter1中配置了全局配置，是一样的这里就不用加了)
scatter2 = (Scatter()
           .add_xaxis([i[0] for i in data[1]])
           .add_yaxis("2015年", [[i[1], i[2], i[3]] for i in data[1]], 
                      symbol_size=JsCode(symbol_js), # 这里配置图形大小，根据GDP总量计算出symbol_size
                      label_opts=opts.LabelOpts(is_show=False),
                      itemstyle_opts=item_style_2)
           )
scatter1.overlap(scatter2)           
scatter1.render_notebook()
```
