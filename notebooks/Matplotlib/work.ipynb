```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

plt.rcParams['font.sans-serif'] = ['SimHei']
plt.rcParams['axes.unicode_minus'] = False
```

```python
df_bank = pd.read_csv('data/ChinaBank.csv', index_col = 'Date')
```

```python
df_bank.index = pd.to_datetime(df_bank.index)
```

```python
df_bank.drop('Unnamed: 0', axis=1, inplace=True)
```

```python
df_bank[['Open','Close']][df_bank.index.year == 2014].plot(figsize = (16,10), rot = 0)
plt.title('2014')
plt.xlabel('')
plt.ylabel('')

plt.legend(labels = ['', ''])
plt.show()
```
