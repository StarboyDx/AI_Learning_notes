## 3. 在图形中加入基本元素并控制

```python
# 1. 准备画布
plt.figure(figsize = (12,8))

# 2. 准备数据
x1 = [1,2,3]
y1 = [4,9,6]

x2 = [1,2,3]
y2 = [20,9,17]

# 3. 图形绘制
plt.plot(x1, y1, 'bo-', label = 'first line')
plt.plot(x2, y2, 'r--*', lw = 4, label = 'second line')

# 4. 添加辅助元素
plt.title('This is a pic\n two lines') # 图形的标题
plt.xlabel('time')
plt.ylabel('number')

# 5. 辅助控制
plt.legend() # 显示图例(plot中的label)
plt.show()
```

```python
# 控制xticks和yticks的标签
# 1. 准备画布
plt.figure(figsize = (12,8))

# 2. 准备数据
x1 = [1,2,3]
y1 = [4,9,6]

x2 = [1,2,3]
y2 = [20,9,17]

# 3. 图形绘制
plt.plot(x1, y1, 'bo-', label = 'first line')
plt.plot(x2, y2, 'r--*', lw = 4, label = 'second line')

# 4. 添加辅助元素
plt.title('This is a pic\n two lines') # 图形的标题
plt.xlabel('time')
plt.ylabel('number')

# 5. 辅助控制
plt.xticks([1,2,3],['Monday','Tuesday','Wednesday'])
plt.yticks([5,10,15,20],['50K','100K','150K','200K'])
# 刻度和标签之间要一一对应

plt.legend() # 显示图例(plot中的label)
plt.show()
```

```python
# 控制坐标轴的范围
# 1. 准备画布
plt.figure(figsize = (12,8))

# 2. 准备数据
x1 = [1,2,3]
y1 = [4,9,6]

x2 = [1,2,3]
y2 = [20,9,17]

# 3. 图形绘制
plt.plot(x1, y1, 'bo-', label = 'first line')
plt.plot(x2, y2, 'r--*', lw = 4, label = 'second line')

# 4. 添加辅助元素
plt.title('This is a pic\n two lines') # 图形的标题
plt.xlabel('time')
plt.ylabel('number')

# 5. 辅助控制
plt.xticks([1,2,3],['Monday','Tuesday','Wednesday'])
plt.yticks([5,10,15,20],['50K','100K','150K','200K'])
# 刻度和标签之间要一一对应
plt.xlim(-1,5)
plt.ylim(0,30)

plt.legend() # 显示图例(plot中的label)
plt.show()
```

```python
# 显示网格线
# 1. 准备画布
plt.figure(figsize = (12,8))

# 2. 准备数据
x1 = [1,2,3]
y1 = [4,9,6]

x2 = [1,2,3]
y2 = [20,9,17]

# 3. 图形绘制
plt.plot(x1, y1, 'bo-', label = 'first line')
plt.plot(x2, y2, 'r--*', lw = 4, label = 'second line')

# 4. 添加辅助元素
plt.title('This is a pic\n two lines') # 图形的标题
plt.xlabel('time')
plt.ylabel('number')

# 5. 辅助控制
plt.xticks([1,2,3],['Monday','Tuesday','Wednesday'])
plt.yticks([5,10,15,20],['50K','100K','150K','200K'])
# 刻度和标签之间要一一对应
plt.xlim(-1,5)
plt.ylim(0,30)

plt.grid()

plt.legend() # 显示图例(plot中的label)
plt.show()
```

```python
# 显示网格线
# 1. 准备画布
plt.figure(figsize = (12,8))

# 2. 准备数据
x1 = [1,2,3]
y1 = [4,9,6]

x2 = [1,2,3]
y2 = [20,9,17]

# 3. 图形绘制
plt.plot(x1, y1, 'bo-', label = 'first line')
plt.plot(x2, y2, 'r--*', lw = 4, label = 'second line')

# 4. 添加辅助元素
plt.title('This is a pic\n two lines') # 图形的标题
plt.xlabel('time')
plt.ylabel('number')

# 5. 辅助控制
# plt.xticks([1,2,3],['Monday','Tuesday','Wednesday'])
# plt.yticks([5,10,15,20],['50K','100K','150K','200K'])
# 刻度和标签之间要一一对应
plt.xlim(-1,5)
plt.ylim(0,30)

plt.grid(axis = 'x', alpha = 0.3) # axis表示在哪个轴上进行网格线的绘制, alpha设置透明度，1是全透明

plt.legend() # 显示图例(plot中的label)
plt.show()
```

```python
# 调整字体效果
# 1. 准备画布
plt.figure(figsize = (12,8))

# 2. 准备数据
x1 = [1,2,3]
y1 = [4,9,6]

x2 = [1,2,3]
y2 = [20,9,17]

# 3. 图形绘制
plt.plot(x1, y1, 'bo-', label = 'first line')
plt.plot(x2, y2, 'r--*', lw = 4, label = 'second line')

# 4. 添加辅助元素
plt.title('This is a pic\n two lines', fontsize = 24, color = 'red') # 图形的标题
plt.xlabel('time', fontsize = 22)
plt.ylabel('number', fontsize = 22)

# 5. 辅助控制
plt.xticks([1,2,3],['Monday','Tuesday','Wednesday'], fontsize = 15)
plt.yticks([5,10,15,20],['50K','100K','150K','200K'], fontsize = 15)
# 刻度和标签之间要一一对应
plt.xlim(-1,5)
plt.ylim(0,30)

plt.grid(axis = 'x', alpha = 0.3) # axis表示在哪个轴上进行网格线的绘制, alpha设置透明度，1是全透明

plt.legend(fontsize = 18) # 显示图例(plot中的label)
plt.show()
```
