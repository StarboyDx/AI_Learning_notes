# 常见图形的绘制
## 折线图

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

plt.rcParams['font.sans-serif'] = ['SimHei']
plt.rcParams['axes.unicode_minus'] = False

# 忽略警告
# import warnings
# warnings.filterwarnings('ignore')
```

普通折线图

```python
# 1.构建画布
plt.figure(figsize = (10, 5))

# 2.数据
x = [10, 20, 30, 50]
y = [40, 100, 200, 150]

# 3.绘制
plt.plot(x, y, 'r*-', label = 'line1')

# 4.辅助对象
plt.title('line plot')

# 5.显示
plt.legend()
plt.show()
```

以时间为横轴的折线图

```python
# 1.构建画布
plt.figure(figsize = (10, 5))

# 2.数据
x = ['2025-09-01', '2025-09-02', '2025-09-03', '2025-09-04']
# 对比下面一行，这里x实际上是对0，1，2，3做了一个映射
y = [40, 100, 200, 150]

# 3.绘制
plt.plot(x, y, 'r*-', label = 'line1')

# 4.辅助对象
plt.title('line plot')
plt.xlabel('date')
plt.ylabel('people')

# 5.显示
plt.legend()
plt.show()
```

## 柱状图

普通柱状图

```python
# 1.构建画布
plt.figure(figsize = (10, 6))

# 2.数据
x = [1, 2, 3, 4]
height = [40, 100, 200, 150]

# 3.绘制
plt.bar(x, height, label = 'bar1')

# 4.辅助对象
plt.title('bar plot')

# 5.显示
plt.legend()
plt.show()
```

非数字的x轴

```python
# 1.构建画布
plt.figure(figsize = (10, 6))

# 2.数据
#x1 = [1, 2, 3, 4]
x = ['cate1', 'cate2', 'cate3', 'cate4']
height = [40, 100, 200, 150]


# 3.绘制
plt.bar(x, height, label = 'bar1', color = '#ffcc00')

# 4.辅助对象
plt.title('bar plot')
plt.xlabel('education')
plt.ylabel('job count')

# 5.显示
plt.legend()
plt.show()
```

多组柱状图

```python
# 1.构建画布
plt.figure(figsize = (10, 6))

# 2.数据
x1 = [1, 2, 3, 4]
#x = ['cate1', 'cate2', 'cate3', 'cate4']
height1 = [40, 100, 200, 150]
x2 = [1, 2, 3, 4]
height2 = [30, 120, 180, 210]

# 3.绘制
plt.bar(x1, height1, label = 'bar1', color = '#ffcc00')
plt.bar(x2, height2, label = 'bar2', color = 'blue', width = 0.5)
# 柱状图如果绘制在同一x坐标上，会产生重叠的效果

# 4.辅助对象
plt.title('bar plot')
# plt.xlabel('education')
# plt.ylabel('job count')

# 5.显示
plt.legend()
plt.show()
```

## 饼图

```python
# 1. 创建画布（方形）
plt.figure(figsize = (12, 12))

# 2. 数据准备
data = [100, 50, 76, 35]
#绘制饼图时会自动换算成百分比
d_labels = ['SEO', 'QQ', 'BBS', 'TMALL'] #渠道
d_colors = ['c', 'm', 'y', 'r'] #颜色

# 3. 绘制图形
plt.pie(
    data,
    labels = d_labels,
    colors = d_colors,
    shadow = True,
    explode = (0, 0.2, 0, 0.1), #控制某个扇区朝圆周方向移动的距离
    autopct = '%.1f%%',
    startangle = 90,
    textprops = {'fontsize': 17}
)

# 4. 辅助对象
plt.title('pie', fontsize = 25)

# 5. 显示隐藏
plt.legend()
plt.show()
```

# 面向对象的图形绘制 ※※※

```python
# 1. 准备工作
# 1.1 画布
fig1 = plt.figure(figsize = (12, 8))
# print(fig1) # 调试用
# 1.2 axes对象
ax1 = plt.subplot(1, 1, 1)
# 在画布上切分为1行1列，并返回第一个区域，返回一个axes对象
#print(ax1) # 调试
#print(type(ax1)) # 调试

# 2. 准备数据
# x,y,width
x_bar = np.arange(4) # 0-3的数组，x轴的坐标
h_bar = np.array([15, 20, 18, 25]) #柱体的高度，y轴的坐标
w_bar = 0.5 # 柱体的宽度

# 3. 图形的绘制
# 3.1 用axes对象来绘制柱状图，目的是返回绘制好的柱体的集合
rects = ax1.bar(
    x = x_bar,
    height = h_bar,
    width = w_bar,
    color = 'lightblue',
    edgecolor = 'blue',
    lw = 3, #line-width
)
# print(rects) # 调试用
# 用axes对象的bar方法绘制柱状图，返回绘制好的柱体集合

# 3.2 遍历柱体集合
for rect in rects:
    #   print(rect)
    # 遍历每个矩形获取坐标和长宽的信息
    x = rect.get_x() # 返回矩形的x坐标
    y = rect.get_height() # 返回矩形的高度
    # 4. 数据标签的绘制
    print(x, y) # 该坐标对应矩形的左上角
    # ax1.plot(x, y, 'ro', ms = 10)
    ax1.plot(x + w_bar/2, y, 'r*', ms = 10) 
    #提取出的坐标，在x轴的位置上加上柱体宽度的一半，就是矩形上方中心点的位置，也就是用来绘制数据标签的位置
    ax1.text(
        x + 0.15,
        y + 0.8,
        str(y) + 'w',
        fontsize = 20,
        color = 'red'
    )
    # 在指定的位置绘制数据标签

# 5. 其他辅助对象处理和设置
# 5.1 调整y轴的坐标
# plt.ylim(0, 28) # 用plt的方法
ax1.set_ylim(0, 28) # 用axes的方法实现
# 5.2 xticklabels的处理
# 5.2.1 限制xtick的显示范围
ax1.set_xticks(x_bar)
ax1.set_xticklabels(['Monday', 'Tuesday', 'Wednesday', 'Thursday'], fontsize = 20, color = 'orange')
# 5.3 yticklabels的处理：不修改内容，只改变文本大小
ax1.set_yticklabels(ax1.get_yticks(), fontsize = 20, color = 'black')
# 5.4 xlabel和ylabel
ax1.set_xlabel('Week', fontsize = 22)
ax1.set_ylabel('Money', fontsize = 22)
# 5.5 title
ax1.set_title('Weekday Sales', fontsize = 26, color = 'brown', y = 1.02)
# y是调整title到图形上边缘的距离，1是紧贴上边缘
# 5.5 网格线
ax1.grid(alpha = 0.5)

# 6. 显示隐藏
plt.show()
```
