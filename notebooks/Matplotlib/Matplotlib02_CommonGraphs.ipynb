# 常见图形的绘制
## 折线图

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

plt.rcParams['font.sans-serif'] = ['SimHei']
plt.rcParams['axes.unicode_minus'] = False
```

普通折线图

```python
# 1.构建画布
plt.figure(figsize = (10, 5))

# 2.数据
x = [10, 20, 30, 50]
y = [40, 100, 200, 150]

# 3.绘制
plt.plot(x, y, 'r*-', label = 'line1')

# 4.辅助对象
plt.title('line plot')

# 5.显示
plt.legend()
plt.show()
```

以时间为横轴的折线图

```python
# 1.构建画布
plt.figure(figsize = (10, 5))

# 2.数据
x = ['2025-09-01', '2025-09-02', '2025-09-03', '2025-09-04']
# 对比下面一行，这里x实际上是对0，1，2，3做了一个映射
y = [40, 100, 200, 150]

# 3.绘制
plt.plot(x, y, 'r*-', label = 'line1')

# 4.辅助对象
plt.title('line plot')
plt.xlabel('date')
plt.ylabel('people')

# 5.显示
plt.legend()
plt.show()
```

```python
# 1.构建画布
plt.figure(figsize = (10, 5))

# 2.数据
#x = ['2025-09-01', '2025-09-02', '2025-09-03', '2025-09-04']
y = [40, 100, 200, 150]

# 3.绘制
plt.plot(y, 'r*-', label = 'line1')
# 省略[x]时，默认生成0开始的整数序列，和y构成坐标点绘制折线图

# 4.辅助对象
plt.title('line plot')

# 5.显示
plt.legend()
plt.show()
```

## 柱状图

普通柱状图

```python
# 1.构建画布
plt.figure(figsize = (10, 6))

# 2.数据
x = [1, 2, 3, 4]
height = [40, 100, 200, 150]

# 3.绘制
plt.bar(x, height, label = 'bar1')

# 4.辅助对象
plt.title('bar plot')

# 5.显示
plt.legend()
plt.show()
```
