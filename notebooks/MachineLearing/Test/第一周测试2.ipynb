```python
import pandas as pd
%matplotlib inline
import matplotlib.pyplot as plt
```

```python
employee = pd.read_csv('data/employee.csv')
```

```python
employee.head()
```

```python
pd.crosstab(index = employee.Gender,
           columns = employee.Jobcat)
```

```python
relaTab = pd.crosstab(index = employee.Gender,
           columns = employee.Jobcat)
```

```python
relaTab.plot(kind = 'bar', stacked = True)
plt.title('relaTab')
plt.xlabel('Gender')
plt.ylabel('Number')
plt.show()
```
