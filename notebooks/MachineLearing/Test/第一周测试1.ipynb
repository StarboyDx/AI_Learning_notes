```python
import numpy as np
import pandas as pd
```

```python
Age = np.random.randint(12,100,40)
```

```python
Age
```

```python
#自定义分段
pd.cut(Age, bins = [12, 20, 40, 60, 100])
```

```python
pd.cut(Age, bins = [12, 20, 40, 60, 100]).value_counts()
```

```python
#等宽分段
pd.cut(Age, bins = 4)
```

```python
pd.cut(Age, bins = 4).value_counts()
```

```python
#等深分段
pd.qcut(Age, q = 4)
```

```python
pd.qcut(Age, q = 4).value_counts()
```
