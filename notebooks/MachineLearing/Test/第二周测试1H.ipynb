```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import pearsonr

plt.rcParams['font.sans-serif'] = ['SimHei']
plt.rcParams['axes.unicode_minus'] = False
```

```python
df = pd.read_csv('data/ad.csv')
df.head()
```

```python
# 计算皮尔逊相关系数
sales = df['sales']
ad_cost = df['ad_cost']

# 计算皮尔逊相关系数r和p值（p值用于检验相关性是否显著，p<0.05则显著）
corr_coef, p_value = pearsonr(sales, ad_cost)
print(f"皮尔逊相关系数 r = {corr_coef:.4f}")
print(f"显著性p值 = {p_value:.6f}")
```

r = 0.9645 >= 0.8 > 0    强正相关，且相关性显著

```python
fig, ax = plt.subplots(figsize=(10, 6))
ax.scatter(ad_cost, sales, s=80, alpha=0.8)
```
