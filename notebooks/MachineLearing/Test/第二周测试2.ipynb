```python
import pandas as pd
from sklearn.preprocessing import StandardScaler, MinMaxScaler

# 1. 构建原始数据
df = pd.DataFrame([['Kelly, Mr. James', 'male', 'Q',25],
                   ['Ilieff, Mr. Ylio', 'male', 'S',36],
                   ['Connolly, Miss. Kate', 'female', 'Q',55],
                   ['Keane, Mr. Daniel', 'male', 'Q',45],
                   ['Assaf, Mr. Gerios', 'male', 'C',18]],
                   columns=['name', 'sex', 'Embarked', 'age'])
```

```python
# 2. 标准化处理
scaler_standard = StandardScaler()
df['age_standardized'] = scaler_standard.fit_transform(df[['age']])

# 3. 归一化处理
scaler_minmax = MinMaxScaler()
df['age_normalized'] = scaler_minmax.fit_transform(df[['age']])

df.head()
```

```python
df_encoded = pd.get_dummies(df, columns=['sex', 'Embarked'])
df_encoded
```
