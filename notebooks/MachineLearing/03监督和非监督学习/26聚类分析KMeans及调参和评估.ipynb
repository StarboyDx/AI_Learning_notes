```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
from sklearn.cluster import KMeans

%matplotlib inline

from IPython.core.interactiveshell import InteractiveShell
InteractiveShell.ast_node_interactivity = 'all'
```

# Read the iris data and visualize it

```python
data = load_iris()
features = data['data']
feature_names = data['feature_names']
target = data['target']

# 利用第1,2个特征作为xy轴作散点图，使用target的值对散点作为标记图例
for t, marker, col in zip(range(3), ">ox", "rgb"):
    plt.scatter(features[target == t, 0], features[target == t, 1], marker = marker, c = col)

plt.xlabel(feature_names[0])
plt.ylabel(feature_names[1])
```

# KMeans Model

```python
kmeans = KMeans(n_clusters = 3)
kmeans.fit(features)

# output result
kmeans.cluster_centers_
kmeans.labels_
```

# Model prediction

```python
kmeans.fit_predict(features)

cluster_label = kmeans.predict(features)
cluster_label
```

# Use the elbow method to tune parameters -- select the most appropriate k value

```python
# 肘部法的核心代码工作：计算每一个K值对应的损失函数
# 我们计算K值从1到10对应的平均畸变程度：

# 用scipy求解距离
from scipy.spatial.distance import cdist

K = range(1,10)
mean_distoritions = []

for k in K:
    kmeans = KMeans(n_clusters = k)
    kmeans.fit(features)
    mean_distoritions.append(sum(np.min(cdist(features, kmeans.cluster_centers_, 'euclidean'), axis = 1)) / features.shape[0])

plt.plot(K, mean_distoritions, 'bx-')
plt.xlabel('k')
plt.ylabel('loss')
plt.title('find the best k value')
```

# Evaluation of Clustering Models -- Silhouette Coefficient

```python
# 轮廓系数
from sklearn.metrics import silhouette_score

silhouette_score(features, kmeans.labels_, metric = 'euclidean')
```

```python
# 看不同K下的结果
K = range(2, 11)
ss = []
for k in K:
    kmeans = KMeans(n_clusters = k)
    kmeans.fit(features)
    ss.append(silhouette_score(features, kmeans.labels_, metric = 'euclidean'))

plt.plot(K, ss, 'bx-')
plt.xlabel('k')
plt.ylabel('silhouette_score')
plt.title('find the maximum silhouette_score')
```
