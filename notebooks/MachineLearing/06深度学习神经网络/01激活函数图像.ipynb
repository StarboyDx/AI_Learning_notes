```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def make_plot(z, f, df, func_name):
    """
    绘制函数曲线及其导数曲线
    参数：
        z: x轴数据（自变量）
        f: 函数值（因变量）
        df: 函数的导数值
        func_name: 函数名称（用于标题/图例）
    """
    plt.figure(figsize=(10, 6))  # 设置画布大小
    # 绘制原函数曲线
    plt.plot(z, f, label=f'{func_name} function', color='blue', linewidth=2)
    # 绘制导数曲线
    plt.plot(z, df, label=f'{func_name} derivative', color='red', linestyle='--', linewidth=2)
    
    # 图表美化
    plt.title(f'{func_name.capitalize()} Function and Its Derivative')
    plt.xlabel('z')
    plt.ylabel('Value')
    plt.grid(True, alpha=0.3)  # 显示网格
    plt.legend()  # 显示图例
    plt.axhline(y=0, color='k', linestyle='-', alpha=0.1)  # x轴参考线
    plt.axvline(x=0, color='k', linestyle='-', alpha=0.1)  # y轴参考线
    plt.show()  # 显示图表
```

```python
z = np.arange(-10, 10, 0.01)
f = 1/(1 + np.exp(-z))
df = f * (1 - f)
make_plot(z, f, df, 'sigmoid')
```

```python
z = np.arange(-10, 10, 0.01)
f = np.tanh(z)
df = 1 - f*f
make_plot(z, f, df, 'tanh')
```

```python
z = np.arange(-10, 10, 0.01)
f = z * (z > 0)
df = 1. * (z > 0)
make_plot(z, f, df, 'ReLU')
```

```python
z = np.arange(-10, 10, 0.01)
f = np.where(z > 0, z, z * 0.01)
df = np.where(z > 0, 1, 0.01)
make_plot(z, f, df, 'Leaky_ReLU')
```
