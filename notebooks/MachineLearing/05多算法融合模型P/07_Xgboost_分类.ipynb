```python
import xgboost as xgb
import pandas as pd
from IPython.core.interactiveshell import InteractiveShell
InteractiveShell.ast_node_interactivity = 'all'
```

```python
train_src = pd.read_csv('../02数据预备_清洗及特征工程/data/titanic.csv')

x = train_src[['Pclass', 'Sex', 'Age', 'SibSp', 'Parch', 'Fare']]
x['Sex'] = train_src['Sex'].replace({'male': 1, 'female':0})
y = train_src['Survived']

x.head()
y.head()
```

```python
from sklearn import model_selection
X_train, X_test, y_train, y_test = model_selection.train_test_split(x, y, test_size = 0.3, random_state = 123456)
```

# Prepare data

```python
# Convert data to xgboost format
dtrain = xgb.DMatrix(X_train, label = y_train)
dtest = xgb.DMatrix(X_test, label = y_test)
```

# Parameters

```python
params = {
    'booster': 'gbtree',
    'objective': 'binary:logistic',
    'early_stopping_rounds': 50,
    'eval_metric': 'auc',
    'gamma': 0,
    'max_depth': 5,
    'subsample': 0.6,
    'colsample_bytree': 0.9,
    'min_child_weight': 1,
    'eta': 0.02,
    'seed': 123456,
    'nthread': 3,
    'silent': 0,
}
```

# Xgboost training model

```python
watchlist = [(dtest, 'val'), (dtrain, 'train')]
model = xgb.train(params, dtrain, num_boost_round = 200, evals = watchlist, early_stopping_rounds = 10)
```

# Make predictions on the test set

```python
# 动态判断XGBoost版本 Dynamically determine the XGBoost version
if xgb.__version__ >= '1.3.0':
    # 新版：用best_iteration + iteration_range
    predict_y = model.predict(dtest, iteration_range = (0, model.best_iteration))
else:
    # 旧版：用原来的best_ntree_limit + ntree_limit
    predict_y = model.predict(dtest, ntree_limit=model.best_ntree_limit) 
# The prediction result is a probability.
predict_y_label = (predict_y > 0.5) * 1 # Convert the predicted probabilities into 0,1 labels

predict_y[:50]
predict_y_label[:50]
```

# Evaluating model performance: AUC, Precision, Accuracy, Recall

```python
from sklearn import metrics

print("AUC: %f" % metrics.roc_auc_score(y_test, predict_y))

print("Precision: %f" % metrics.precision_score(y_test, predict_y_label))
print("Recall: %f" % metrics.recall_score(y_test, predict_y_label))
print("Accuracy: %f" % metrics.accuracy_score(y_test, predict_y_label))
```
