```python
from sklearn import datasets

# import iris datasets
iris = datasets.load_iris()
X = iris.data
y = iris.target
```

```python
X[:10]
y[:10]
```

LDA Dimensionality Reduction

```python
# from sklearn.lda import LDA
# The package path has been changed in the latest version; please use the following line of code to import it:
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis as LDA

# define
lda = LDA(n_components=2)

lda.fit(X, y) # y: target
X_new = lda.transform(X)
X_new[:5]

#fit_transform
X_new = lda.fit_transform(X, y)
X_new[:5]

#predict(X)
lda.predict(X)
#score
lda.score(X, y)
```

Comparison of Linear Discriminant Analysis (LDA) and Principal Component Analysis (PCA)

```python
from sklearn.decomposition import PCA

pca = PCA(n_components = 2)
pca.fit(X)
X_new = pca.transform(X)

import matplotlib.pyplot as plt
%matplotlib inline

plt.scatter(X_new[:, 0], X_new[:, 1], marker = 'o', c = y)
plt.show()
```

```python
lda = LDA(n_components=2)
X_new = lda.fit_transform(X, y)

plt.scatter(X_new[:, 0], X_new[:, 1], marker = 'o', c = y)
plt.show()
```
