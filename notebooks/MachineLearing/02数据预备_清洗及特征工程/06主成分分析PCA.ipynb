# Iris Datasets

```python
from sklearn import datasets

# import iris data
iris = datasets.load_iris()
X = iris.data
y = iris.target

X[:10]
y[:10]
```

# PCA Dimensionality Reduction

```python
from sklearn.decomposition import PCA

#sklearn.decomposition.PCA(n_components = None, copy = True, whiten = False)
#n_components: 主成分个数
#copy: 表示是否再运行算法时，将原始训练数据复制一份，省略默认为True
#whiten: 白化，使得每个特征具有相同的方差，缺省时默认为False

#note: 下面注释总结下目前sklearn构建模型基本步骤
#先用相应类定义一个模型，这里定义一个PCA模型
pca = PCA(n_components = 3)

#用fit方法拟合这个特征模型
pca.fit(X)

#利用模型做转换，这里用X数据集训练，因此也用X来转换
X_new = pca.transform(X)
X_new[:5]

#以上两个步骤可以使用fit_transform合并
X_new = pca.fit_transform(X)
X_new[:5]

#输出主成分解释方差占比
print(pca.explained_variance_ratio_)
print(pca.explained_variance_)
```

# Visualization after PCA Dimensionality Reduction

```python
# 降到两个维度比较适合可视化
pca = PCA(n_components = 2)
pca.fit(X)
X_new = pca.transform(X)

import matplotlib.pyplot as plt
%matplotlib inline

plt.scatter(X_new[:, 0], X_new[:, 1], marker = 'o', c = y)
plt.show()
```
