```python
import pandas as pd
```

```python
titanic = pd.read_csv('data/titanic.csv')
titanic.info()
```

# Fill in Missing Values

```python
## Fill in Missing Values of Age with Mean Value
age_mean = round(titanic['Age'].mean()) # round 四舍五入
titanic['Age'].fillna(age_mean, inplace = True)
titanic.info()
titanic.head()
```

# Construct Indicator Variables for Missing Values (0/1)

```python
titanic = pd.read_csv('data/titanic.csv')
titanic.info()

titanic['Age_ismissing'] = 0
titanic.loc[titanic['Age'].isnull(), 'Age_ismissing'] = 1
titanic['Age_ismissing'].value_counts()
```
