# The indexing and slicing of arrays.  
## Indexing and slicing of one-dimensional arrays (refer to lists)

```python
a1 = np.arange(9)
a1
```

```python
a1[-1]
```

```python
a1[1:-1]
```

```python
a1[-1::-1]
```

## Multiple dimensional arrays

```python
a2 = np.arange(9).reshape(3,3)
```

```python
a2
```

```python
a2[1,2] #Nested list cannot be accessed using multidimensional indexing
```

```python
a3 = np.random.randint(1,100,27).reshape((3,3,3))
```

```python
a3
```

```python
a3[1,1,2]
```

```python
a3[1]
```

```python
a4 = np.arange(1,36).reshape(5,7)
```

```python
a4
```

```python
a4[0:2,0:]
```

```python
a4[:2] #Slice the rows directly
```

```python
a4[5:] #Attention: When slicing only the final dimension, leading dimensions must remain intact
```

```python
a4[:,5:]
```

```python
#reverse
a4[::-1,::-1]
```

```python
a4[2:4,2:4] = 0
a4
```

### Magic index  
Using arrays/lists as indices of arrays: Access by position

```python
x1 = np.arange(10)
```

```python
x1
```

```python
x1[[3,4,7]]
```

```python
index1 = np.array([[0,2],[1,3]])
```

```python
x1[index1]
```

```python
x2 = np.arange(20).reshape((4,5))
```

```python
x2
```

```python
x2[0::2]
```

```python
x2[[0,2]]
```

```python
x2[:,[1,3,4]]
```

```python
x2[[0,2,3],[0,1,4]]
```

### Boolean indexing
Access by condition

```python
#Find elements greater than 4
```

```python
x1[x1>4]
```

```python
x1 > 4
```

```python
x1
```

```python
#Convert numbers greater than 4 in x1 to 1, and numbers less than or equal to 4 to 0
#(discretization of continuous variables)
x1[x1<=4] = 0
x1[x1>4] = 1
```

```python
x1
```

```python
x2 > 5
```

```python
x2[x2>5]
```

```python
x2[:,3][x2[:,3]>5]
```

```python
#Compound condition
x2[:,3][(x2[:,3] % 2 == 1) | (x2[:,3] > 10)]
```
