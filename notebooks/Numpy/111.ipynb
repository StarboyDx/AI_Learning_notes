#Numpy的导入：针对矩阵/向量的运算 ----  Numpy import： For matrix and vector computations
##需求：矩阵相加
        [0, 1, 4] + [0, 1, 8] = [0, 2, 12]

```python
import numpy as np
```

```python
def npSum(n):
    a = np.arange(n) ** 2
    b = np.arange(n) ** 3
    return a+b
```

```python
npSum(3)
```

```python
npSum(1000)[-3:]
```

```python
%timeit npSum(1000)
```

```python
def npSumBig(n):
    a = np.arange(n, dtype = np.int64) ** 2
    b = np.arange(n, dtype = np.int64) ** 3
    return a+b
```

```python
npSumBig(1000)[-3:]
```
