## Solving systems of linear equations
鸡兔同笼 35个头，94条腿\
1x + 1y = 35\
2x + 4y = 94\
np.linalg.solve(系数矩阵，结果矩阵）

```python
#Coefficient matrix
arr1 = np.array([[1,1],[2,4]])
```

```python
arr2 = np.array([35,94])
```

```python
np.linalg.solve(arr1,arr2)
```
