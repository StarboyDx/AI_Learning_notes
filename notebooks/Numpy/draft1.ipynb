```python
a = np.array([[1,2,3],[3,4,5],[4,5,6]])

print(a) # 从某个索引处开始切割

print('从数组索引 a[1:] 处开始切割')

print(a[1:])
```

```python
a = np.arange(9)
```

```python
print ('第一个数组：')
print (a) 
print ('\n')
```

```python
print ('将数组分为三个大小相等的子数组：')

b = np.split(a,3)

print (b)

print ('\n')
```

```python
print ('将数组在一维数组中表明的位置分割：')

b = np.split(a,[4,7])

print (b)
```

```python
a = np.array([0.25, 1.33, 1, 100])

print ('我们的数组是：')

print (a)

print ('\n')

print ('调用 reciprocal 函数：')

print (np.reciprocal(a))
```
