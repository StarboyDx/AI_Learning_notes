```python
#Construct a function that converts dates string to the days of the week.
def dstr2weekday(datestr):
    return datetime.strptime(datestr, '%d-%m-%Y').date().isoweekday()
```

```python
dstr2weekday(eDate[0])
```

```python
Weekday2,Close2 = np.loadtxt('Datebase/data.csv',
                             delimiter = ',',
                             skiprows = 1,
                             usecols = (1, 6),
                             unpack = True,
                             encoding = 'GBK',
                             converters = {1:dstr2weekday}
)
```

```python
Weekday2
```

```python
eWeek1
```

```python
Close2
```

```python
#Construct an empty array to store the average closing prices from Monday to Friday.
avgs = np.zeros(5)
```

```python
avgs
```

```python
for i in range(1, 6):
    #Iterate through Monday to Friday
    indexarray = np.where(Weekday2 == i)
    print(indexarray)
    price = np.take(Close2, indexarray)
    print(price)
    avg = np.mean(price)
    print(avg)
    avgs[i-1] = avg
```

```python
print(avgs)
```

```python
print('max : ', np.max(avgs), 'day : ', np.argmax(avgs) + 1)
```
