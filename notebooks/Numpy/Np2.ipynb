# Array splitting
## horizental splitting

```python
as1 = np.arange(9).reshape((3,3))
as1
```

```python
as2 = np.arange(9).reshape((3,-1))
as2
```

```python
re1 = np.hsplit(as1,3)
re1
```

```python
type(re1)
```

```python
for item in re1:
    print(item)
    print(item.shape)
```

```python
re1a = np.split(as1,3,axis = 1)
re1a
```

## Vertical splitting

```python
re3 =  np.vsplit(as1,3)
re3
```

```python
for item in re3:
    print(item)
    print(item.shape)
```

```python
re3a = np.split(as1,3)
re3a
```

## Depth splitting

```python
as2 = np.arange(64).reshape(4,4,4)
as2
```

```python
re5 = np.dsplit(as2,2)
re5
```
