```python
import pandas as pd
import numpy as np
```

```python
true_data = np.array(['B','B','B','B','B','B','B','B','B','B',
                      'B','B','B','B','B','B','B','M','M','B',
                      'B','M','M','M','B','B','M','M','B','M',
                      'B','M','M','M','B','B','M','M','B','M',
                      'B','M','M','M','B','B','M','M','B','M',
                      'M','B','B','B','B','B','B'])
pred_data = np.array(['B']*36 + ['M']*2 + ['M']*17 + ['B']*2)
```

```python
cross_tab = pd.crosstab(true_data, pred_data, rownames=['确诊'], colnames=['预测'], margins=True)
```

```python
data = {
    "HAND": ["right", "left", "right", "left", "left"],
    "sex": ["男", "女", "男", "男", "女"],
    "iq": [61, 99, 94, 60, 52],
    "eq": [51, 52, 91, 67, 84]
}
df = pd.DataFrame(data)
```

```python
pivot = pd.pivot_table(
    df,
    index="HAND",
    columns="sex",
    values=["iq", "eq"],
    aggfunc="mean"
)
```
