```python
import pandas as pd
```

```python
df = pd.read_excel('work6-3.xlsx', sheet_name='Sheet1', skiprows = [0])
```

```python
df
```

```python
CustomerSales = df.groupby('客户名称')['销量'].sum().reset_index()
```

```python
CustomerSales
```

```python
CustomerSales = CustomerSales.sort_values('销量', ascending = False).reset_index(drop = True)
```

```python
CustomerSales
```

```python
def GetImportance(rank):
    if rank <= 2: #前两名
        return '* * *'
    elif 3 <= rank <=5:
        return '* *'
    else:
        return '*'
```

```python
CustomerSales['重要程度'] = CustomerSales.index.map(lambda x: GetImportance(x+1))
```

```python
CustomerSales
```

```python
result = CustomerSales[['重要程度','客户名称', '销量']]
```
