## Bank dataset analysis and pivot tables

```python
df_bank = pd.read_csv('Database/data/bank-full.csv', sep = ';')
```

```python
df_bank.head()
```

```python
df_bank.info()
```

```python
df_bank.describe()
```

```python
df_bank[['age']].describe()
```

```python
df_bank.job.describe()
```

```python
df_bank.groupby('job').age.count()
```

```python
type(df_bank.groupby('job').age.count())
```

```python
df_bank.groupby('job').age.count().sort_values(ascending = False)
```

```python
?pd.Series.value_counts
```

```python
df_bank.education.value_counts()
```

```python
# The impact of occupation on purchase decisions
df_bank.groupby(['job', 'y']).age.count()
```

```python
df_bank.groupby(['job', 'y'])['age'].count().unstack()
```

```python
# The average age of people with different educational backgrounds and marital statuses
df_bank.groupby(['education', 'marital']).age.mean().unstack()
```

```python
# 直接构建 更直观 相当于excel中的透视表操作
df_bank.pivot_table(index = ['education'], columns = ['marital'], values = ['age'], aggfunc = ['mean'])
```

```python
 # aggfunc default: mean
df_bank.pivot_table(index=['job'], values = ['age', 'duration'])
```

```python
df_bank.pivot_table(index=['job', 'education'], 
                    columns = ['marital'],
                    values = ['age', 'duration'],
                   aggfunc = ['mean', 'max'])
```
