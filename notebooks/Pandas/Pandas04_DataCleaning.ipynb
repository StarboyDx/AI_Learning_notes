# Concatenating datasets
## merge

```python
from pandas import Series,DataFrame
```

```python
dfa = DataFrame(
    {
        '学号' : ['#001', '#002', '#003'],
        '语文' : [87,98,67],
        '数学' : [92,86,78],
        '英语' : [100,87,98]
    },
    index = [1,5,4])
```

```python
dfb = DataFrame(
    {
        '学号' : ['#001', '#002', '#003'],
        '体育' : [83,68,87]
    })
```

```python
dfb
```

```python
pd.merge(dfa,dfb)
```

```python
df_m1 = pd.merge(dfa,dfb)
```

```python
df_m1 #indexs changed
```

```python
dfc = DataFrame(
    {
        '学号' : ['#001', '#002', '#004'],
        '体育' : [83,68,77]
    })
```

```python
df_m2 = pd.merge(dfa, dfc, on = '学号', how = 'inner') 
```

```python
df_m2
```

```python
df_m3 = pd.merge(dfa, dfc, on = '学号', how = 'left') 
```

```python
df_m3
```

```python
df_m4 = pd.merge(dfa, dfc, on = '学号', how = 'right') 
```

```python
df_m4
```

```python
df_m5 = pd.merge(dfa, dfc, on = '学号', how = 'outer')
```

```python
df_m5
```
